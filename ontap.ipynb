```python
gpa = float(input("nhap gpa: " ))
if gpa >= 8.5:
    print("xuat sac")
elif gpa >= 7:
    print("gioi")
elif gpa >= 5.5:
    print("trung binh")
else :
    print ("yeu")
```

```python
n = int(input("nhap so nguyen duong: "))
tong = 0
for i in range (1, n+1):
  if i % 2 == 0 :
    tong += i
    print(tong)

```

```python
password = "python123"
count = 0

while count < 3:
    nhap = input("Nhập mật khẩu: ")
    if nhap == password:
        print("Đăng nhập thành công")
        break
    else:
        print("Sai mật khẩu")
        count += 1

if count == 3:
    print("Tài khoản bị khóa")
```
